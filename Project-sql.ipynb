```python
import psycopg2
from psycopg2 import Error
from psycopg2.extras import NamedTupleCursor

def execute_query(query, fetch_result=False):
    try:
        connection = psycopg2.connect(
                        database="project_cake_store", 
                        user='postgres',
                        password='123', 
                        host='localhost', 
                        port='5432'
                    )
        
        connection.autocommit = True
        cursor = connection.cursor(cursor_factory=NamedTupleCursor)
        cursor.execute(query)
        if fetch_result:
            return cursor.fetchall()
    except (Exception, Error) as error:
        print("Error while connecting to PostgreSQL", error)
    finally:
        if (connection):
            cursor.close()
            connection.close()
```

```python
#Найти дату последнего заказа
query = """SELECT max(making_date)
           FROM orders
           
        """
all_rows = execute_query(query, fetch_result=True)
for row, value in enumerate(all_rows):
    print(row, value)
```

```python
#Информация по последнему заказу
query = """SELECT *
           FROM orders
           ORDER BY making_date DESC 
           limit 1
           
        """
all_rows = execute_query(query, fetch_result=True)
for row, value in enumerate(all_rows):
    print(row, value)
```

```python
#Самая популярная начинка
query = """SELECT filling.name, COUNT(filling.name) as order_count
           FROM filling
           JOIN cake ON filling.id = cake.type_id
           GROUP BY filling.name 
           ORDER BY order_count DESC
           limit 1
           
        """
all_rows = execute_query(query, fetch_result=True)
for row, value in enumerate(all_rows):
    print(row, value)
```

```python
#Выручка на этой неделе
query = """SELECT SUM(total_cost) FROM cake
           JOIN orders ON orders.cake_id = cake.id
           WHERE making_date >= to_timestamp('19-12-2022', 'dd-mm-yyyy') AND making_date <= to_timestamp('26-12-2022', 'dd-mm-yyyy')
        """
all_rows = execute_query(query, fetch_result=True)
for row, value in enumerate(all_rows):
    print(row, value)
```

```python
#Выручка на прошлой неделе
query = """SELECT SUM(total_cost) FROM cake
           WHERE id IN (SELECT cake_id FROM orders WHERE making_date BETWEEN to_timestamp('12-12-2022', 'dd-mm-yyyy') AND to_timestamp('19-12-2022', 'dd-mm-yyyy'))
        """
all_rows = execute_query(query, fetch_result=True)
for row, value in enumerate(all_rows):
    print(row, value)
```

```python
#Выручка за последнюю неделю
query = """SELECT SUM(total_cost) FROM cake
           JOIN orders ON orders.cake_id = cake.id
           WHERE making_date >= now() - interval '7 days'
        """

all_rows = execute_query(query, fetch_result=True)
for row, value in enumerate(all_rows):
    print(row, value)
```
